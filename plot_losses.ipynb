```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List
```

```python
def plot_loss(loss: List[float], title: str, save: str = None):
  plt.plot(range(len(loss)), loss)
  plt.title(title)
  plt.xlabel('Iteration')
  plt.ylabel('Loss')
  plt.show()
  if save:
    plt.savefig(save)
```

```python
lstm_losses = np.load('losses/lstm_losses.npy')
plot_loss(lstm_losses, 'LSTM Training Loss', save='plots/lstm_training_loss.png')
```

```python
kgram_losses = np.load('losses/kgram_losses2.npy')
plot_loss(lstm_losses, 'KGram Training Loss', save='plots/kgram_training_loss.png')
```
